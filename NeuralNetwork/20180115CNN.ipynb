```python
import numpy as np 
np.random.seed(1337) # For reproducibility
from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation, Convolution2D, MaxPooling2D, Flatten
from keras.optimizers import Adam
import csv
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
import numpy as np 
import pickle
import os
import cv2
```

```python
def reshape_image(img, num):
	img = Image.open(img)
	a, b = img.size[0], img.size[1]
	if a > b:
		scale = a - b 
		img = img.crop((scale / 2, 0, a - scale / 2, b))
		c = b
	else:
		scale = b - a
		img = img.crop((0, scale / 2, a, b - scale / 2))
		c = a

	img = img.resize((num, num), Image.ANTIALIAS)

	return img
```

```python
pixel = 80
covl = 16
pooln = 2
fli = 64

learn_rate = 1e-4
epoch = 2
```

```python
Y = []
for i in range(273*3):
    Y.append(0)
for i in range(148*3):
    Y.append(1)
print(len(Y))

```

```python
X_train, X_test, y_train, y_test = train_test_split(data, Y, test_size = 100)

# Data pre-processing
y_train = np_utils.to_categorical(y_train, num_classes = 2)
y_test = np_utils.to_categorical(y_test, num_classes = 2)



```

```python
# Build CNN
model = Sequential()

# Conv layer 1 output shape(32, 28, 28)
model.add(Convolution2D(
	filters = fli, kernel_size = (covl, covl), padding = "same",   # Padding method
	input_shape = (1, pixel, pixel)
	))
model.add(Activation("relu"))
# Pooling layer 1 (max pooling) output shape (32, 14, 14)
model.add(MaxPooling2D(
	pool_size = pooln, strides = (2, 2), padding = "same" # Padding method
	))

# Conv layer 2 output shape (64, 14, 14)
model.add(Convolution2D(
	filters = fli, kernel_size = (covl, covl), padding = "same",   # Padding method
	))

model.add(Activation("relu"))

# Pooling layer 2 (Max pooling) output shape(64, 7, 7)
model.add(MaxPooling2D(
	pool_size = pooln, strides = (2, 2), padding = "same" # Padding method
	))

# Fully connected layer 1 input shape (64 * 7 * 7) = (3136), output shape (1024)
model.add(Flatten())
model.add(Dense(500))
model.add(Activation("relu"))
model.add(Dense(2))
model.add(Activation("softmax"))

# Define optimizer
adam = Adam(lr = learn_rate)

# Add metrics to get more information
model.compile(optimizer = adam, loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
print("\nTraining -------------")
model.fit(X_train, y_train, epochs = epoch, batch_size = 64)

print("\nTesting  -------------")
loss, accuracy = model.evaluate(X_test, y_test)

print("\ntest loss:", loss)
print("\ntest accuracy:", accuracy)



model.save("/Users/pengbohao/Downloads/my_model.h5")

```
